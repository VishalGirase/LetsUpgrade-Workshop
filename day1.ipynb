```python
def profit_or_loss(cost_prize,selling_prize):
    print(cost_prize)
#     print(selling_prize)
    if selling_prize > cost_prize:
        print(selling_prize,"",'\t Profit')
    if cost_prize > selling_prize:
        print(selling_prize,"","\t Loss")
    if selling_prize == cost_prize:
        print(selling_prize,"",'\t Neither')
```

```python
profit_or_loss(20,30)
```
